```
# 下載 APK
from google.colab import files
import glob
apk_files = glob.glob('bin/*.apk')
if apk_files:
    for apk in apk_files:
        print(f'✅ {apk}')
        files.download(apk)
else:
    print('❌ APK 未找到')
    print('查看 .buildozer/logs/ 目錄:')
    !ls -la .buildozer/logs/ 2>/dev/null || echo '無日誌目錄'
```
