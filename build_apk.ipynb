# Kivy Android APK Builder (Google Colab)

這個筆記本可以幫助您在 Google Colab 的免費 Linux 環境中編譯 Kivy App。

### 使用步驟：
1. 點擊上方的「執行階段」->「全部執行」
2. 等待環境安裝完成
3. 當出現上傳按鈕時，上傳您的 `buildozer.spec` 和 `main.py` 等專案檔案 (或直接從 GitHub 下載)
4. 等待編譯完成，最後會自動下載 APK 檔案

```python
# 1. 安裝必要依賴
!sudo apt-get update
!sudo apt-get install -y build-essential libsqlite3-dev sqlite3 bzip2 libbz2-dev zlib1g-dev libssl-dev openssl libgdbm-dev libgdbm-compat-dev liblzma-dev libreadline-dev libncursesw5-dev libffi-dev uuid-dev libltdl-dev autoconf automake libtool
!sudo apt-get install -y libffi-dev
!sudo apt-get install -y python3-pip python3-setuptools
!sudo apt-get install -y openjdk-17-jdk

# 2. 安裝 Buildozer
!pip install buildozer cython==0.29.36
```

```python
# 3. 下載您的專案代碼 (從 GitHub)
!git clone https://github.com/yungtang20/twse-app.git
%cd twse-app

# 確保 buildozer.spec 存在
!ls -la
```

```python
# 4. 開始編譯 (這需要一些時間，約 15-30 分鐘)
# 同意授權條款
!yes | buildozer android debug
```

```python
# 5. 下載 APK
from google.colab import files
import glob

apk_files = glob.glob("bin/*.apk")
if apk_files:
    for apk in apk_files:
        print(f"Found APK: {apk}")
        files.download(apk)
else:
    print("No APK found. Build might have failed.")
```
