```
# 下載 APK
from google.colab import files
import glob
apk_files = glob.glob('bin/*.apk')
if apk_files:
    for apk in apk_files:
        print(f'✅ {apk}')
        files.download(apk)
else:
    print('❌ APK 未找到')
```
